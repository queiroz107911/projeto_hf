## 3) Liste os nomes das colunas e descreva brevemente o que você entende que cada coluna representa.

```python
df.columns
```

Código da Emenda: identificador único de cada emenda parlamentar registrada no orçamento.

Ano da Emenda: ano em que a emenda foi apresentada ou incluída no orçamento.

Tipo de Emenda: classificação da emenda, como individual, de bancada ou de relator.

Código do Autor da Emenda: código identificador do parlamentar responsável pela emenda.

Nome do Autor da Emenda: nome do deputado ou senador que propôs a emenda.

Número da emenda: número específico que identifica a emenda dentro do processo orçamentário.

Localidade de aplicação do recurso: local onde os recursos da emenda serão aplicados.

Código Município IBGE: código oficial do município segundo a base do IBGE.

Município: nome do município que receberá ou onde será aplicado o recurso.

Código UF IBGE: código da unidade federativa (estado) segundo o IBGE.

UF: sigla do estado brasileiro onde o recurso será aplicado.

Região: região do Brasil à qual o estado pertence (Norte, Nordeste, Centro-Oeste, Sudeste ou Sul).

Código Função: código que representa a função orçamentária do gasto público.

Nome Função: nome da função orçamentária, como saúde, educação ou transporte.

Código Subfunção: código que detalha a subfunção dentro da função orçamentária.

Nome Subfunção: nome da subfunção que especifica melhor a área do gasto público.

Código Programa: código do programa governamental relacionado ao recurso.

Nome Programa: nome do programa governamental ao qual a emenda está vinculada.

Código Ação: código da ação orçamentária específica dentro do programa.

Nome Ação: descrição da ação governamental que receberá os recursos.

Código Plano Orçamentário: código do plano orçamentário associado à ação.

Nome Plano Orçamentário: nome ou descrição do plano orçamentário.

Valor Empenhado: valor do recurso que foi oficialmente reservado pelo governo para pagamento.

Valor Liquidado: valor do recurso que já teve a despesa confirmada após a entrega do serviço ou produto.

Valor Pago: valor que efetivamente já foi pago pelo governo.

Valor Restos A Pagar Inscritos: valores empenhados em anos anteriores que ainda não foram pagos e foram transferidos para pagamento futuro.

Valor Restos A Pagar Cancelados: parte dos restos a pagar que foi cancelada e não será mais paga.

Valor Restos A Pagar Pagos: valores de restos a pagar que finalmente foram pagos em anos posteriores.
